```python
%matplotlib inline

import os
from os.path import join
import numpy as np
import keras
from keras import layers
from tensorflow import data as tf_data
import matplotlib.pyplot as plt
import struct
import random
from array import array
from IPython.display import display, Markdown
import triangle_gen
from sklearn.model_selection import train_test_split
```

# model

```python
dataset_size = 10000
triangles = []

x_data_set = []
y_data_set = []



for n in range(dataset_size):
    #2 angles and 1 sides known
    triangle = triangle_gen.TriangleGenerator().generate_random_triangle_via_angles(known_indices=[0,2,4])
    x = (triangle.angles[0], triangle.angles[2], triangle.sides[0], triangle.triangle_type)
    y = (triangle.sides[2], triangle.sides[0], triangle.angles[1])
    x_data_set.append(x)
    y_data_set.append(y)
    
x_data_set = np.array(x_data_set)
y_data_set = np.array(y_data_set)
    
print(x_data_set)
print(y_data_set)
X_train, X_test, y_train, y_test = train_test_split(x_data_set, y_data_set, test_size=0.2, random_state=42)


model_1 = keras.Sequential([
    keras.layers.Flatten(input_shape=[4]),        # In layer
    keras.layers.Dense(128, activation=keras.activations.relu),        # Hidden layer
    keras.layers.Dense(128, activation=keras.activations.linear),        # Hidden layer
    keras.layers.Dense(128, activation=keras.activations.relu),        # Hidden layer
    keras.layers.Dense(3, activation='linear')       # Out layer
])

model_1.compile(optimizer='adam', loss='mean_squared_error')

model_1.fit(X_train, y_train, epochs=50, batch_size=8)

score = model_1.evaluate(X_test, y_test, verbose=0)
print("Accuracy:", score)

display(Markdown(f"Accuracy: {score}<br>"))
```
