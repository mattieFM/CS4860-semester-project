# model

```python
def generic_run(func, label):
    dataset_size = 1000
    triangles = []

    x_data_set = []
    y_data_set = []



    for n in range(dataset_size):
        x = random.random()*1000
        y = func(x)
        x_data_set.append(x)
        y_data_set.append(y)
        
    x_data_set = np.array(x_data_set)
    y_data_set = np.array(y_data_set)
        
    X_train, X_test, y_train, y_test = train_test_split(x_data_set, y_data_set, test_size=0.2, random_state=42)


    model_1 = keras.Sequential([
        keras.layers.Flatten(input_shape=[1]),        # In layer
        keras.layers.Dense(128, activation=keras.activations.relu),        # Hidden layer
        keras.layers.Dense(128, activation=keras.activations.linear),        # Hidden layer
        keras.layers.Dense(128, activation=keras.activations.relu),        # Hidden layer
        keras.layers.Dense(1, activation=keras.activations.linear)       # Out layer
    ])

    model_1.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])

    model_1.fit(X_train, y_train, epochs=50, batch_size=8)

    score = model_1.evaluate(X_test, y_test, verbose=1)
    display(Markdown(f"### Model Name: '{label}' Statistics:<br>"))
    display(Markdown(f"mse: {score[1]}<br>  Loss: {score[0]}<br>"))
    
generic_run(math.sin, "sin")
generic_run(math.cos, "cos")
generic_run(math.tan, "tan")
generic_run(lambda x:x+10, "x+10")
```
